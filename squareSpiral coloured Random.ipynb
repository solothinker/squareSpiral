# Random Coloured Square Spiral

```python
%matplotlib nbagg
import numpy as np
import cv2
import matplotlib.pyplot as plt
import random
```

```python
# R = 256
# G = 256
# B = 256
pix = 256
nAxis = int(np.sqrt(pix*pix*pix))
low = 0
high = nAxis-1
count = int((nAxis+1)/2)
random.seed(1)
```

<a href="https://www.youtube.com/watch?v=3hLlxWDQZcI">Reference Link</a>

```python
%%time
cList =[]
for ii in range(count):
    for jj in range(low,high+1,1): # top index
        cList.append((ii,jj))
    for jj in range(low+1,high+1,1): # last column
        cList.append((jj,high))
    for jj in range(high-1,low-1,-1):
        cList.append((high,jj))
    for jj in range(high-1,low,-1):
        cList.append((jj,low))
    low += 1
    high -= 1    
```

```python
R = np.zeros((nAxis,nAxis),dtype=np.uint8)
G = np.zeros((nAxis,nAxis),dtype=np.uint8)
B = np.zeros((nAxis,nAxis),dtype=np.uint8) 
```

```python
%%time
indx = 0
for r in range(pix):
    for g in range(pix):
        for b in range(pix):
            R[cList[indx]] = random.randint(0,r)
            G[cList[indx]] = random.randint(0,g)
            B[cList[indx]] = random.randint(0,r)
            indx += 1
```

```python
%%time
img_bgr = cv2.merge([R,G,B])
plt.figure(figsize = (5,5))
plt.imshow(img_bgr)
plt.axis('off')
# plt.savefig('RGB.jpg',dpi=80)
plt.show()
```

```python
%%time
img_bgr = cv2.merge([R,B,G])
plt.figure(figsize = (5,5))
plt.imshow(img_bgr)
plt.axis('off')
# plt.savefig('RBG.jpg',dpi=80)
plt.show()
```
